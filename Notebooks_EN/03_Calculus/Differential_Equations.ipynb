# Verify if $ \psi(t, x) = A \cos(\omega t + kx) $ satisfies the wave equation

---

### Problem Statement:
We want to check if the function:
$$
\psi(t, x) = A \cos(\omega t + kx)
$$
is a solution to the wave equation:
$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0
$$
where $ v = \frac{\omega}{k} = \frac{2\pi / T}{2\pi / \lambda} $.

---

### Step 1: Compute $ \frac{\partial^2 \psi}{\partial t^2} $

The given function is:
$$
\psi(t, x) = A \cos(\omega t + kx)
$$

First, compute the first partial derivative with respect to $ t $:
$$
\frac{\partial \psi}{\partial t} = -A \omega \sin(\omega t + kx)
$$

Now, compute the second partial derivative with respect to $ t $:
$$
\frac{\partial^2 \psi}{\partial t^2} = -A \omega^2 \cos(\omega t + kx)
$$

---

### Step 2: Compute $ \frac{\partial^2 \psi}{\partial x^2} $

First, compute the first partial derivative with respect to $ x $:
$$
\frac{\partial \psi}{\partial x} = -A k \sin(\omega t + kx)
$$

Now, compute the second partial derivative with respect to $ x $:
$$
\frac{\partial^2 \psi}{\partial x^2} = -A k^2 \cos(\omega t + kx)
$$

---

### Step 3: Substitute into the wave equation

The wave equation is:
$$
\frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = 0
$$

Substitute $ \frac{\partial^2 \psi}{\partial t^2} = -A \omega^2 \cos(\omega t + kx) $ and $ \frac{\partial^2 \psi}{\partial x^2} = -A k^2 \cos(\omega t + kx) $:
$$
- A \omega^2 \cos(\omega t + kx) - v^2 (- A k^2 \cos(\omega t + kx)) = 0
$$

Simplify:
$$
- A \omega^2 \cos(\omega t + kx) + v^2 A k^2 \cos(\omega t + kx) = 0
$$

Factor out $ A \cos(\omega t + kx) $ (assuming $ A \neq 0 $ and $ \cos(\omega t + kx) \neq 0 $):
$$
-A \cos(\omega t + kx) \left( \omega^2 - v^2 k^2 \right) = 0
$$

---

### Step 4: Verify the condition $ \omega^2 = v^2 k^2 $

From the definition of the wave speed $ v = \frac{\omega}{k} $, we know:
$$
v^2 = \frac{\omega^2}{k^2}
$$

Thus:
$$
\omega^2 = v^2 k^2
$$

This satisfies the equation:
$$
-A \cos(\omega t + kx) \cdot 0 = 0
$$

---

### Conclusion:

The function $ \psi(t, x) = A \cos(\omega t + kx) $ is a solution to the wave equation:
$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0
$$
