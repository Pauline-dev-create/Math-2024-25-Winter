#: Second-Order Surfaces

## General Equations of Second-Order Surfaces

### 1. **Ellipsoid**

An ellipsoid is a three-dimensional analogue of an ellipse. The general equation of an ellipsoid centered at \((h, k, l)\) is:

$\frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} + \frac{(z - l)^2}{c^2} = 1$

Where:
- \(a, b, c\) are the semi-axes of the ellipsoid along the \(x\), \(y\), and \(z\) axes respectively.

For an ellipsoid centered at the origin \((0, 0, 0)\), the equation simplifies to:

$\frac{x^2}{a^2} + \frac{y^2}{b^2} + \frac{z^2}{c^2} = 1$

#### Special Case: Sphere
If \(a = b = c = r\), the ellipsoid becomes a sphere with the equation:

$(x - h)^2 + (y - k)^2 + (z - l)^2 = r^2$

For a sphere centered at the origin \((0, 0, 0)\):


$x^2 + y^2 + z^2 = r^2$

---

### 2. **Hyperboloid**

#### One-Sheet Hyperboloid
The general equation of a hyperboloid of one sheet is:


$\frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} - \frac{(z - l)^2}{c^2} = 1$

This represents a hyperboloid with a connected surface.

#### Two-Sheet Hyperboloid
The general equation of a hyperboloid of two sheets is:


$\frac{(x - h)^2}{a^2} - \frac{(y - k)^2}{b^2} - \frac{(z - l)^2}{c^2} = 1$

This represents a hyperboloid with two disconnected surfaces.

For hyperboloids centered at the origin \((0, 0, 0)\), the equations simplify to:

- One-sheet hyperboloid:
$\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = 1$
- Two-sheet hyperboloid:

$\frac{x^2}{a^2} - \frac{y^2}{b^2} - \frac{z^2}{c^2} = 1$

## Rotational Surfaces

Rotational surfaces are generated by rotating a curve about an axis.

### 1. **Surface of Revolution**
A surface of revolution is created by rotating a curve \(y = f(x)\) about the \(z\)-axis. Its equation is:

$x^2 + y^2 = [f(z)]^2$

### 2. **Rotational Ellipsoid**
Rotating an ellipse about one of its axes generates a rotational ellipsoid. For example, if the ellipse:

$\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$

is rotated about the \(z\)-axis, the equation of the surface becomes:

$\frac{x^2}{a^2} + \frac{y^2}{a^2} + \frac{z^2}{b^2} = 1$

### 3. **Rotational Hyperboloid**
Rotating a hyperbola about its axis generates a rotational hyperboloid. For example, rotating the hyperbola:

$\frac{x^2}{a^2} - \frac{z^2}{c^2} = 1$

about the \(y\)-axis results in:

$\frac{x^2}{a^2} - \frac{y^2}{b^2} + \frac{z^2}{c^2} = 1$

---

## Applications of Second-Order Surfaces

- **Ellipsoids** are used to model planetary shapes and reflectors in telescopes.
- **Spheres** represent uniform three-dimensional objects like bubbles or planets.
- **Hyperboloids** are used in cooling towers, modern architecture, and engineering designs.
- **Rotational surfaces** describe objects like domes, lenses, and mirrors.
