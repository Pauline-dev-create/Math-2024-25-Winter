# Useful Formulas for Solving Derivative Problems

## 1. **Basic Rules of Differentiation**
- **Constant Rule**:
  $$\frac{d}{dx}[c] = 0 \quad \text{where } c \text{ is a constant.}$$
- **Power Rule**:
  $$\frac{d}{dx}[x^n] = nx^{n-1}.$$
- **Constant Multiple Rule**:
  $$\frac{d}{dx}[cf(x)] = c \cdot f'(x).$$
- **Sum/Difference Rule**:
  $$\frac{d}{dx}[f(x) \pm g(x)] = f'(x) \pm g'(x).$$

## 2. **Product Rule**
If \( y = u(x) \cdot v(x) \), then:
$$\frac{d}{dx}[uv] = u'v + uv'.$$

## 3. **Quotient Rule**
If \( y = \frac{u(x)}{v(x)} \), then:
$$\frac{d}{dx}\left[\frac{u}{v}\right] = \frac{u'v - uv'}{v^2}, \quad v(x) \neq 0.$$

## 4. **Chain Rule**
If \( y = f(g(x)) \), then:
$$\frac{d}{dx}[f(g(x))] = f'(g(x)) \cdot g'(x).$$

## 5. **Derivatives of Common Functions**
- $$\frac{d}{dx}[\sin(x)] = \cos(x).$$
- $$\frac{d}{dx}[\cos(x)] = -\sin(x).$$
- $$\frac{d}{dx}[\tan(x)] = \sec^2(x), \quad x \neq \frac{\pi}{2} + n\pi.$$
- $$\frac{d}{dx}[\sec(x)] = \sec(x)\tan(x).$$
- $$\frac{d}{dx}[\csc(x)] = -\csc(x)\cot(x).$$
- $$\frac{d}{dx}[\ln(x)] = \frac{1}{x}, \quad x > 0.$$
- $$\frac{d}{dx}[\exp(x)] = \exp(x).$$
- $$\frac{d}{dx}[\sqrt{x}] = \frac{1}{2\sqrt{x}}.$$

## 6. **Trigonometric Identities** (Useful for Simplifications)
- $$\sin(2x) = 2\sin(x)\cos(x).$$
- $$1 + \tan^2(x) = \sec^2(x).$$

## 7. **L'Hospital's Rule** (for Improper Limits)
If a limit evaluates to an indeterminate form (\( \frac{0}{0} \) or \( \frac{\infty}{\infty} \)):
$$\lim_{x \to a} \frac{f(x)}{g(x)} = \lim_{x \to a} \frac{f'(x)}{g'(x)},$$
provided the derivatives exist.

## 8. **Second Derivative**
If \( y = f(x) \), then the second derivative is:
$$f''(x) = \frac{d^2y}{dx^2} = \frac{d}{dx}\left(f'(x)\right).$$

## 9. **Applications in Physics**
- **Velocity**:
  $$V(t) = \frac{dx}{dt}.$$
- **Acceleration**:
  $$a(t) = \frac{dV}{dt} = \frac{d^2x}{dt^2}.$$

## 10. **Logarithmic and Exponential Rules**
- $$\frac{d}{dx}[\ln(f(x))] = \frac{f'(x)}{f(x)}, \quad f(x) > 0.$$
- $$\frac{d}{dx}[\exp(f(x))] = \exp(f(x)) \cdot f'(x).$$


# Step-by-Step Solution: Compute Derivatives of Functions

## **1. $ y(x) = -3x + 3 $**
Using the linearity of derivatives:
$$
\frac{dy}{dx} = \frac{d}{dx}(-3x) + \frac{d}{dx}(3) = -3 + 0 = -3.
$$

## **2. $ y(x) = \pi x + \sin(1) $**
\(\pi\) is a constant, and \(\sin(1)\) is also constant:
$$
\frac{dy}{dx} = \frac{d}{dx}(\pi x) + \frac{d}{dx}(\sin(1)) = \pi + 0 = \pi.
$$

## **3. $ y(x) = 4 + \sin(2) $**
Both terms are constants:
$$
\frac{dy}{dx} = \frac{d}{dx}(4) + \frac{d}{dx}(\sin(2)) = 0 + 0 = 0.
$$

## **4. $ y(x) = 2x^3 - 3x^2 + 8x - 9 $**
Using the power rule:
$$
\frac{dy}{dx} = \frac{d}{dx}(2x^3) - \frac{d}{dx}(3x^2) + \frac{d}{dx}(8x) - \frac{d}{dx}(9),
$$
$$
\frac{dy}{dx} = 6x^2 - 6x + 8.
$$

## **5. $ y(x) = 6x^{1/3} $**
Using the power rule:
$$
\frac{dy}{dx} = 6 \cdot \frac{1}{3}x^{-2/3} = 2x^{-2/3}.
$$

## **6. $ y(x) = \sqrt{x} $**
Using \(\sqrt{x} = x^{1/2}\) and the power rule:
$$
\frac{dy}{dx} = \frac{1}{2}x^{-1/2} = \frac{1}{2\sqrt{x}}.
$$

## **7. $ y(x) = \cos(x) + \sin(x) $**
Using the derivatives of trigonometric functions:
$$
\frac{dy}{dx} = -\sin(x) + \cos(x).
$$

## **8. $ y(x) = 2\sin(x)\cos(x) $**
Using the product rule:
$$
\frac{dy}{dx} = 2\left[\frac{d}{dx}(\sin(x))\cos(x) + \sin(x)\frac{d}{dx}(\cos(x))\right],
$$
$$
\frac{dy}{dx} = 2[\cos^2(x) - \sin^2(x)] = 2\cos(2x) \quad (\text{using } \cos^2(x) - \sin^2(x) = \cos(2x)).
$$

## **9. $ y(x) = x\sin(x) $**
Using the product rule:
$$
\frac{dy}{dx} = \frac{d}{dx}(x)\sin(x) + x\frac{d}{dx}(\sin(x)),
$$
$$
\frac{dy}{dx} = \sin(x) + x\cos(x).
$$

## **10. $ y(x) = (x+1)(x+1) $**
Expanding first:
$$
y(x) = (x+1)^2.
$$
Using the chain rule:
$$
\frac{dy}{dx} = 2(x+1)\frac{d}{dx}(x+1) = 2(x+1).
$$

## **11. $ y(x) = \frac{x}{x+1} $**
Using the quotient rule:
$$
\frac{dy}{dx} = \frac{(1)(x+1) - x(1)}{(x+1)^2} = \frac{x+1-x}{(x+1)^2} = \frac{1}{(x+1)^2}.
$$

## **12. $ y(x) = (x+1)\exp(x) $**
Using the product rule:
$$
\frac{dy}{dx} = \frac{d}{dx}(x+1)\exp(x) + (x+1)\frac{d}{dx}(\exp(x)),
$$
$$
\frac{dy}{dx} = \exp(x) + (x+1)\exp(x) = (x+2)\exp(x).
$$

## **13. $ y(x) = \sin(x^2) $**
Using the chain rule:
$$
\frac{dy}{dx} = \cos(x^2)\frac{d}{dx}(x^2) = \cos(x^2) \cdot 2x = 2x\cos(x^2).
$$

## **14. $ y(x) = \exp(-2x) $**
Using the chain rule:
$$
\frac{dy}{dx} = \exp(-2x)\frac{d}{dx}(-2x) = -2\exp(-2x).
$$

## **15. $ y(x) = \frac{1}{\sin(x+1)} $**
Using the chain rule and the reciprocal rule:
$$
\frac{dy}{dx} = -\frac{1}{\sin^2(x+1)}\frac{d}{dx}(\sin(x+1)),
$$
$$
\frac{dy}{dx} = -\frac{\cos(x+1)}{\sin^2(x+1)}.
$$

## **16. $ y(x) = \sqrt{2x+1} $**
Using the chain rule:
$$
\frac{dy}{dx} = \frac{1}{2\sqrt{2x+1}}\frac{d}{dx}(2x+1),
$$
$$
\frac{dy}{dx} = \frac{1}{2\sqrt{2x+1}} \cdot 2 = \frac{1}{\sqrt{2x+1}}.
$$

## **1. Prove: $ \frac{d}{dx} (\ln(\sin(x))) = \cot(x) $**

Let $ y = \ln(\sin(x)) $. Using the chain rule:
$$
\frac{d}{dx}(\ln(\sin(x))) = \frac{1}{\sin(x)} \cdot \frac{d}{dx}(\sin(x)).
$$
Since $ \frac{d}{dx}(\sin(x)) = \cos(x) $, we get:
$$
\frac{d}{dx}(\ln(\sin(x))) = \frac{\cos(x)}{\sin(x)}.
$$
By the definition of cotangent, \( \cot(x) = \frac{\cos(x)}{\sin(x)} \). Hence:
$$
\frac{d}{dx} (\ln(\sin(x))) = \cot(x).
$$

## **2. Verify: For $ f(x) = \cos(x) $, show \( f''(x) = -f(x) $**

Start with \( f(x) = \cos(x) \):
$$
f'(x) = \frac{d}{dx}(\cos(x)) = -\sin(x).
$$
Find the second derivative:
$$
f''(x) = \frac{d}{dx}(-\sin(x)) = -\cos(x).
$$
Since \( f(x) = \cos(x) \), this gives:
$$
f''(x) = -f(x).
$$

## **3. Find Improper Limits Using de l'Hospital's Rule**

### (a) $ \displaystyle \lim_{x \to 0} \frac{\sin(x)}{x} $
This is an indeterminate form \( \frac{0}{0} \). Apply de l'Hospital's Rule:
$$
\lim_{x \to 0} \frac{\sin(x)}{x} = \lim_{x \to 0} \frac{\cos(x)}{1}.
$$
Since \( \cos(0) = 1 \), the limit is:
$$
\lim_{x \to 0} \frac{\sin(x)}{x} = 1.
$$

### (b) $ \displaystyle \lim_{x \to \infty} \frac{\ln(x)}{x} $
This is an indeterminate form $ \frac{\infty}{\infty} $. Apply de l'Hospital's Rule:
$$
\lim_{x \to \infty} \frac{\ln(x)}{x} = \lim_{x \to \infty} \frac{\frac{1}{x}}{1}.
$$
Simplify:
$$
\lim_{x \to \infty} \frac{\ln(x)}{x} = \lim_{x \to \infty} \frac{1}{x} = 0.
$$

### (c) $ \displaystyle \lim_{x \to \infty} \frac{\exp(x)}{x} $
This is an indeterminate form $ \frac{\infty}{\infty} $. Apply de l'Hospital's Rule:
$$
\lim_{x \to \infty} \frac{\exp(x)}{x} = \lim_{x \to \infty} \frac{\exp(x)}{1}.
$$
Since $ \exp(x) \to \infty $ as $ x \to \infty $, the limit is:
$$
\lim_{x \to \infty} \frac{\exp(x)}{x} = \infty.
$$

### (a) Velocity $ V(t) = x'(t) $
Find the derivative:
$$
V(t) = \frac{d}{dt}(3t^2 - 6t + 1) = 6t - 6.
$$

### (b) Acceleration $ a(t) = V'(t) = x''(t) $
Find the derivative of \( V(t) \):
$$
a(t) = \frac{d}{dt}(6t - 6) = 6.
$$

### (c) Evaluate at $ t = 2 $
Substitute \( t = 2 \) into the velocity:
$$
V(2) = 6(2) - 6 = 12 - 6 = 6.
$$
Substitute \( t = 2 \) into the acceleration:
$$
a(2) = 6.
$$

- Acceleration at \( t = 2 \): \( a(2) = 6 \)
