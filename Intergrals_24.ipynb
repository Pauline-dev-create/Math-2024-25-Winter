##1. Basic Integration Rules
1 .Power Rule  
   For $n \neq -1$:  
$ \int x^n \, dx = \frac{x^{n+1}}{n+1} + C$

2.Exponential Functions

$\int e^x \, dx = e^x + C$

3. Trigonometric Functions
  
  $ \int \cos(x) \, dx = \sin(x) + C $

4. Logarithmic Function

   $\int \frac{1}{x} \, dx = \ln|x| + C $

Properties of Integrals
1. Linearity

   $\int [af(x) + bg(x)] \, dx = a\int f(x) \, dx + b\int g(x) \, dx$

2. Definite Integrals  
  
   $\int_a^b f(x) \, dx = F(b) - F(a) $
   where $F(x)$ is the antiderivative of $f(x)$.

Special Techniques
1. Integration by Substitution

   If $u = g(x)$, then:  

  $\int f(g(x))g'(x) \, dx = \int f(u) \, du $

2. Integration by Parts
   
   $\int u \, dv = uv - \int v \, du $
   Choose $u$ and $dv$ carefully.

3. Trigonometric Identities  
   
  $ \sin^2(x) + \cos^2(x) = 1 $
  
   $\cos(2x) = \cos^2(x) - \sin^2(x) $

Area Under a Curve
For a function $f(x)$ over an interval $[a, b]$:  

$\text{Area} = \int_a^b f(x) \, dx $

Length of a Curve
For a function $f(x)$ over an interval $[a, b]$:  

$L = \int_a^b \sqrt{1 + \left(\frac{dy}{dx}\right)^2} \, dx $

Distance Traveled

For a particle with position $x(t)$, the distance traveled is given by:  

$\text{Distance} = \int_a^b |v(t)| \, dt $
where $v(t) = x'(t)$ is the velocity.

$\int (x^2 + 2) \, dx = \int x^2 \, dx + \int 2 \, dx $

$ \int x^2 \, dx = \frac{x^3}{3}, \quad \int 2 \, dx = 2x$

$2 \int \sin(x) \, dx = -2\cos(x) + C$

#$\int \frac{3}{x} \, dx$  

Factor out the constant 3:

$\int \frac{3}{x} \, dx = 3 \int \frac{1}{x} \, dx $

The integral of $\frac{1}{x}$ is $\ln|x|$:

$    \int \frac{3}{x} \, dx = 3\ln|x| + C $

#$\int \frac{1}{x^2} \, dx$  

 Rewrite $\frac{1}{x^2}$ as $x^{-2}$:

$\int \frac{1}{x^2} \, dx = \int x^{-2} \, dx$
Use the power rule:
$\int x^{-2} \, dx = \frac{x^{-1}}{-1} = -\frac{1}{x} + C$

#$\int \left( \frac{1}{3}x^4 - 5 \right) \, dx$  

Split the integral:   $\int \left( \frac{1}{3}x^4 - 5 \right) \, dx = \int\frac{1}{3}x^4 \, dx - \int 5 \, dx$
  
$ \int \frac{1}{3}x^4 \, dx = \frac{1}{3} \cdot \frac{x^5}{5} = \frac{x^5}{15}, \quad \int 5 \, dx = 5x$

Combine results:
  
$\int \left( \frac{1}{3}x^4 - 5 \right) \, dx = \frac{x^5}{15} - 5x + C$

#$\int (5 \sin x + 3e^x) \, dx$  

Split the integral
$ \int (5 \sin x + 3e^x) \, dx = \int 5 \sin x \, dx + \int 3e^x \, dx$

Compute each term:

$\int 5 \sin x \, dx = -5 \cos x, \quad \int 3e^x \, dx = 3e^x$

Combine results:

$\int (5 \sin x + 3e^x) \, dx = -5 \cos x + 3e^x + C$

#$\int \sqrt[3]{x} \, dx$

$\int \sqrt[3]{x} \, dx = \int x^{1/3} \, dx$

Use the power rule:

$\int x^{1/3} \, dx = \frac{x^{4/3}}{4/3} = \frac{3}{4}x^{4/3} + C$

#$\int \sqrt{10x} \, dx$

$\int \sqrt{10x} \, dx = \sqrt{10} \int x^{1/2} \, dx$

 Use the power rule:

$\int x^{1/2} \, dx = \frac{x^{3/2}}{3/2} = \frac{2}{3}x^{3/2}$

Combine results:
$\int \sqrt{10x} \, dx = \frac{2\sqrt{10}}{3}x^{3/2} + C$

 Substitute back $u$:

$\int \cos\left(\frac{5}{2}x + 3\right) \, dx = \frac{2}{5} \sin\left(\frac{5}{2}x + 3\right) + C$

##3. Area of the region bounded by $x = 1$, $x = 2$, $y = 0$, and $y = x^2 + 1$}

The area is given by:
$$
A = \int_1^2 (x^2 + 1) \, dx
$$
Split the integral:
$$
\int_1^2 (x^2 + 1) \, dx = \int_1^2 x^2 \, dx + \int_1^2 1 \, dx
$$
Compute each term:
$$
\int_1^2 x^2 \, dx = \left[ \frac{x^3}{3} \right]_1^2 = \frac{2^3}{3} - \frac{1^3}{3} = \frac{8}{3} - \frac{1}{3} = \frac{7}{3}
$$
$$
\int_1^2 1 \, dx = [x]_1^2 = 2 - 1 = 1
$$
Combine results:
$$
A = \frac{7}{3} + 1 = \frac{10}{3}
$$

##4. Area under the sine curve over $[0, \pi]$}

The area is given by:
$$
P = \int_0^\pi \sin(x) \, dx
$$
The integral of $\sin(x)$ is $-\cos(x)$:
$$
P = \left[ -\cos(x) \right]_0^\pi = -\cos(\pi) + \cos(0)
$$
Evaluate:
$$
-\cos(\pi) + \cos(0) = -(-1) + 1 = 2
$$
So:
$$
P = 2
$$
